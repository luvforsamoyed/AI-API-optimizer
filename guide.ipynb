```python
#existing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json, datetime
```

```python
#custom library
import sampling
import noAPI
import azure
```

```python
#generate virtual dataset
abn_pt, y = sampling.value_modeling(sig = 10)
dt = sampling.gen_time_idx(dtype = 'datetime')
ts = sampling.gen_time_idx(dtype = 'timestamp')
```

```python
#visualize virtual dataset
req_stamp = pd.Series(y, index = dt)
req_stamp.plot(x=req_stamp.index, y=req_stamp.values, figsize=(12,6))
plt.xlabel('Date time')
plt.ylabel('Value')
plt.title("Virtual Time Series")
```

```python
#visualize virtual dataset
req_stamp = pd.Series(y, index = dt)
req_stamp.plot(x=req_stamp.index, y=req_stamp.values, figsize=(12,6))
for i in abn_pt:
    plt.axvline(x=i, linewidth = 1, color = 'r')
plt.xlabel('Date time')
plt.ylabel('Value')
plt.title("Virtual Time Series")
```

```python
noAPI_f1 = noAPI.f1_metrics(abn_pt, y, ts)
print("F1 score for noAPI is: %f" %noAPI_f1)
```

```python
azure_f1 = azure.f1_metrics(abn_pt, y, dt)
print("F1 score for Azure is: %f" %azure_f1)
```
